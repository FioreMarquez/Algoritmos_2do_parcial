### Ejercicio 1.1

Implementar una clase `CuentaBancaria`.

La clase debe tener los siguientes atributos privados:

- `__numero`: número de cuenta (`int`)
- `__titular`: nombre del titular (`str`)
- `__saldo`: saldo disponible (`float`)

- Constructor Implementar:

  ```python
  def __init__(self, numero: int, titular: str, saldo_inicial: float = 0.0):
    ...
  ```
  Con los siguientes requisitos:

    - `numero` debe ser un entero positivo.
    - `titular` debe ser un string no vacío.
    - `saldo_inicial` debe ser un número mayor o igual a 0.
    - Los atributos deben inicializarse como privados.

- Métodos getters

  - ```python
    def get_numero(self) -> int:
      ...
    ```
  - ```python
    def get_titular(self) -> str:
      ...
    ```
  - ```python
    def get_saldo(self) -> float:
      ...
    ```

- Métodos modificadores

    - ```python
        def depositar(self, monto: float) -> None:
        ...
      ```

        Requisitos:

        - `monto` debe ser un número positivo.
        - Debe sumar el monto al saldo.

    - ```python
        def extraer(self, monto: float) -> None:
        ...
      ```
        Requisitos:

        - `monto` debe ser un número positivo.
        - Debe fallar con assert si no hay saldo suficiente.

- Método especial
  ```python
  def __str__(self) -> str:
      ...
  ```

  Debe devolver un string con el siguiente formato:
  ```python
  "Cuenta #1001 | Titular: Ana Pérez | Saldo: $1500.0"
  ```

Comportamiento deseado de la clase:
```python
c = CuentaBancaria(1001, "Ana Pérez", 1000)

assert c.get_numero() == 1001
assert c.get_titular() == "Ana Pérez"
assert c.get_saldo() == 1000

c.depositar(500)
assert c.get_saldo() == 1500

c.extraer(300)
assert c.get_saldo() == 1200

print(c)
```

````python
# codigo
class CuentaBancaria:
    def __init__(self, numero, titular, saldo_inicial):
        assert type(numero)== int and numero >0, ' debe ingresar un numero entero positivo'
        assert type(titular) == str and len(titular)>0, ' debe ingresar una cadena de caracteres positiva'
        assert isinstance(saldo_inicial, (int, float)) and saldo_inicial >=0, ' debe ingresar un numero mayor o igual a cero'

        self.__numero = numero
        self.__titular = titular
        self.__saldo_inicial = saldo_inicial

    def despositar(self, monto:float):
        assert isinstance(monto, (int, float)) and monto > 0, ' debe ingresar un monto positivo'
        return monto + self.__saldo_inicial
    
    def extraer(self, monto):
        assert isinstance(monto, (int, float)) and monto > 0
        print('no hay saldo suficiente')
    def __str__(self):
        return f"Cuenta: {self.__numero}| Titular: {self.__titular}| Saldo: {self.__saldo_inicial}"
````
