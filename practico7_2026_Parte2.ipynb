## CSV

CSV (Comma Separated Values) es un formato simple para guardar datos en filas y columnas.
Ejemplo:
```python
nombre,edad,ciudad
Juan,30,Córdoba
María,25,Rosario
```

**Ideas importantes**
- Cada línea representa un registro.
- Los datos están separados por comas (u otro delimitador).
- Es un archivo de texto plano.
- Muy usado para exportar datos desde Excel o bases de datos.

1. Leer un CSV, aqui cada fila es una lista.
    ```python
    import csv

    with open("datos.csv", newline="") as archivo:
        reader = csv.reader(archivo)

        for fila in reader:
            print(fila)
    ```
2. `next(reader)` se usa para saltear la primer fila de encabezados.

3.  Escribir un CSV
    ```python
    import csv

    with open("datos.csv", "w", newline="") as archivo:
        writer = csv.writer(archivo)

        writer.writerow(["nombre", "edad"])
        writer.writerow(["Ana", 20])
    ```

| Funcionalidad        | Descripción                            | Ejemplo                              |
| -------------------- | -------------------------------------- | ------------------------------------ |
| `import csv`         | Importa el módulo CSV                  | `import csv`                         |
| `csv.reader()`       | Lee un archivo CSV fila por fila       | `lector = csv.reader(archivo)`       |
| `next(lector)`       | Saltea el encabezado                   | `next(lector)`                       |
| `for fila in lector` | Recorre las filas del CSV              | `for fila in lector:`                |
| `fila[i]`            | Accede a una columna por índice        | `fila[0]`                            |
| `csv.DictReader()`   | Lee filas como diccionarios            | `lector = csv.DictReader(archivo)`   |
| `fila["columna"]`    | Accede a una columna por nombre        | `fila["edad"]`                       |
| `csv.writer()`       | Permite escribir archivos CSV          | `escritor = csv.writer(archivo)`     |
| `writerow()`         | Escribe una sola fila                  | `escritor.writerow(["Ana", 20])`     |
| `writerows()`        | Escribe varias filas                   | `escritor.writerows(datos)`          |

## JSON

JSON es un formato muy usado para intercambio de datos y APIs. Se parece mucho a los diccionarios y listas de Python. Ejemplo:

```python
{
    "nombre": "Juan",
    "edad": 30
}
```

1. Leer JSON
    ```python
    import json

    with open("archivo.json") as f:
        texto = f.read()

        datos = json.loads(texto)

        print(datos)
    ```
    `json.loads()` convierte texto JSON a estructuras Python.

2. Convertir Python → JSON
    ```python
    import json

    diccionario = {"a": 1, "b": 2}

    texto_json = json.dumps(diccionario)

    print(texto_json)
    ```
    `json.dumps()` convierte estructuras Python a texto JSON.

    **Cosas importantes**: Conversión de tipos
    - None  -> null
    - True  -> true
    - False -> false
    - Las claves numéricas pasan a strings.

3. JSON desde Internet
    ```python
    import requests

    url = "https://ejemplo.com/datos.json"

    response = requests.get(url)

    if response.status_code == 200:
        datos = response.json()
    ```
    `response.json()` ya convierte automáticamente el JSON a Python.

## XML

XML es un formato jerárquico basado en etiquetas. Ejemplo:
```python
<inventario>
  <producto codigo="P001">
    <nombre>Teclado Mecánico</nombre>
    <marca>Logitech</marca>
    <precio moneda="USD">85.00</precio>
  </producto>
  
  <producto codigo="P002">
    <nombre>Monitor 4K</nombre>
    <marca>Dell</marca>
    <precio moneda="EUR">320.00</precio>
  </producto>
</inventario>
```
Conceptos importantes
- Elementos: inventario, producto, nombre, marca y precio.
- Atributos: codigo (para identificar el producto) y moneda (para especificar el tipo de divisa del precio)

1. Leer XML
    ```python
    import xml.dom.minidom

    # Cargamos el archivo
    doc = xml.dom.minidom.parse("archivo.xml")

    # Obtenemos la lista de productos
    productos = doc.getElementsByTagName("producto")

    for p in productos:
        # 1. Leer un atributo del elemento padre (producto)
        codigo = p.getAttribute("codigo")
        
        # 2. Leer el texto de los elementos hijos
        nombre = p.getElementsByTagName("nombre")[0].firstChild.data
        
        # 3. Leer un atributo de un elemento hijo (precio)
        precio_elem = p.getElementsByTagName("precio")[0]
        precio_valor = precio_elem.firstChild.data
        moneda = precio_elem.getAttribute("moneda")
        
        print(f"Código: {codigo} | Producto: {nombre} | Precio: {precio_valor} {moneda}")
    ```
    - `getAttribute("nombre_atributo")`: Es la función específica para obtener el valor que pusiste dentro de las etiquetas.
    - `firstChild.data`: Se usa para acceder al contenido de texto que está "atrapado" entre las etiquetas <nombre>...</nombre>.
    - Índice [0]: `getElementsByTagName` siempre devuelve una lista, por lo que usamos [0] para tomar el primer (y normalmente único) elemento encontrado.

| Funcionalidad                  | Descripción                             | Ejemplo                               |
| ------------------------------ | --------------------------------------- | ------------------------------------- |
| `from xml.dom import minidom`  | Importa el módulo para trabajar con XML | `from xml.dom import minidom`         |
